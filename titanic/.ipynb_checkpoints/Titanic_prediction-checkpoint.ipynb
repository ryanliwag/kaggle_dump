```python
#Import Dataset
train_dataset = pd.read_csv("dataset/train.csv")
test_dataset = pd.read_csv("dataset/test.csv")
```

```python
test = train_dataset
y_train = test["Survived"]
test=test.drop(["Survived"], axis=1)
```

```python
combined = test.append(test_dataset)
```

```python
combined.reset_index().drop(["index"], axis=1)
combined = combined.drop(["PassengerId"], axis=1)
```

```python
#process Sex
combined['Sex'].replace(to_replace=dict(female=0, male=1), inplace=True)
```

```python
#process_Pclass():
combined.Pclass.isnull().any()
one_hot_pclass= pd.DataFrame(pd.get_dummies(combined["Pclass"]))
combined=pd.concat([combined, one_hot_pclass], axis=1)
```

```python
#passenger names

combined=combined.reset_index().drop(["index"], axis=1)
```

```python
combined['title'] = combined['Name'].map(lambda x: x.split(',')[1].split('.')[0].strip())
combined.title.unique
```

```python

# Passenger classes

#process_Embarked

Title_Dictionary = {
    "Capt": "Officer",
    "Col": "Officer",
    "Major": "Officer",
    "Jonkheer": "Royalty",
    "Don": "Royalty",
    "Sir" : "Royalty",
    "Dr": "Officer",
    "Rev": "Officer",
    "the Countess":"Royalty",
    "Mme": "Mrs",
    "Mlle": "Miss",
    "Ms": "Mrs",
    "Mr" : "Mr",
    "Mrs" : "Mrs",
    "Miss" : "Miss",
    "Master" : "Master",
    "Lady" : "Royalty"
}

combined['title'] = combined.title.map(Title_Dictionary)
one_hot_title = pd.DataFrame(pd.get_dummies(combined["title"]))
combined = pd.concat([combined, one_hot_title], axis=1)
```

```python
combined
```

```python
#process age

grouped_train = combined.iloc[:891].groupby(['Sex','Pclass','title'])
grouped_median_train = grouped_train.median()
grouped_median_train = grouped_median_train.reset_index()[['Sex', 'Pclass', 'title', 'Age']]
```

```python
def fill_age(row):
    condition = (
        (grouped_median_train['Sex'] == row['Sex']) & 
        (grouped_median_train['title'] == row['title']) & 
        (grouped_median_train['Pclass'] == row['Pclass'])
    ) 
    return grouped_median_train[condition]['Age'].values[0]


def process_age():
    global combined
    # a function that fills the missing values of the Age variable
    combined['Age'] = combined.apply(lambda row: fill_age(row) if np.isnan(row['Age']) else row['Age'], axis=1)
    return combined
```

```python
combined = process_age()
```

```python
# process embarked 

combined.Embarked.value_counts()
combined.Embarked.fillna('S', inplace=True)

one_hot_embarked = pd.get_dummies(combined["Embarked"], prefix="Embarked_")
combined = pd.concat([combined, one_hot_embarked], axis=1)
```
